# how many total number of days does the flights table cover?
    

```python

total_days = 0
for month in flights['month'].unique():
    total_days += flights[flights['month']== month]['day'].unique()[-1]
print(total_days)
```

# how many departure cities (not airports) does the flights database cover?

```python
airports.head()
```

```python
lst = []
for origin in flights['origin'].unique():
    if airports[airports['IATA_CODE']== origin]['CITY'].unique() not in lst:
        lst.append(airports[airports['IATA_CODE']== origin]['CITY'].unique())
print(len(lst))
```
